```python
import glob
import re

# Get a list of all CSV file paths in the current directory
csv_files = glob.glob("/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/records/*.csv")

# Extract date and time information from each file name
file_info = []
for file_path in csv_files:
    file_name = file_path.split("/")[-1]  # Extract file name from the file path

    # Use regular expressions to extract date and time from the file name
    match = re.search(r'_(\d{4}-\d{2}-\d{2})_\[(\d{2}-\d{2}-\d{2})\]', file_name)
    if match:
        date = match.group(1)
        time = match.group(2)
        file_info.append((file_name, date, time))

# Sort file names based on the date and time format mentioned in their names
sorted_files = sorted(file_info, key=lambda x: (x[1], x[2]))

# Print the sorted file names
for file_info in sorted_files:
    print(file_info[0])
```

```python
import glob
import shutil
import os
import pandas as pd

# Get a list of all CSV file paths in the current directory
csv_files = glob.glob("/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/records/*.csv")

# Create a new folder to move the files with '#'
destination_folder = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/records/records_with_timestamp"
os.makedirs(destination_folder, exist_ok=True)

# Extract date and time information from each file name and group based on the '#' column
files_with_hash_column = []
files_without_hash_column = []

for file_path in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Check if the '#' column exists in the DataFrame
    if '#' in df.columns:
        files_with_hash_column.append(file_path)
    else:
        files_without_hash_column.append(file_path)

# Move files with '#' column to the new folder
for file_path in files_with_hash_column:
    file_name = os.path.basename(file_path)
    shutil.move(file_path, os.path.join(destination_folder, file_name))

# Print the groups
print("Files without '#' column:")
for file_path in files_without_hash_column:
    print(file_path)

print("\nFiles with '#' column moved to:", destination_folder)
```

```python
import os
import pandas as pd
import numpy as np

# Define input and output folders
input_folder = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/updated_records_2/"
output_folder = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/processed_records/"

# Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Iterate over each file in the input folder
for file_name in os.listdir(input_folder):
    if file_name.endswith('.csv'):
        input_file_path = os.path.join(input_folder, file_name)
        output_file_path = os.path.join(output_folder, file_name)

        # Read the CSV file
        df = pd.read_csv(input_file_path)

        # Convert degrees to radians for latitude and longitude
        df['latitude_radians'] = np.radians(df['OSD.latitude'])
        df['longitude_radians'] = np.radians(df['OSD.longitude'])

        # Calculate rate of change of latitude, longitude, and height in degrees per second
        df['rate_of_change_lat_mrcodegree'] = df['latitude_radians'].diff() * (360 / (2 * np.pi))* 1e6  / df['OSD.flyTime [s]'].diff()
        df['rate_of_change_long_mrcodegree'] = df['longitude_radians'].diff() * (360 / (2 * np.pi))* 1e6 / df['OSD.flyTime [s]'].diff()
        df['rate_of_change_height'] = df['OSD.height [m]'].diff() / df['OSD.flyTime [s]'].diff()

        # Calculate time difference
        df['time_diff'] = df['OSD.flyTime [s]'].diff()

        # Handle cases where time difference is zero
        zero_time_diff_indices = df[df['time_diff'] == 0].index
        df.loc[zero_time_diff_indices, ['rate_of_change_lat_mrcodegree', 'rate_of_change_long_mrcodegree', 'rate_of_change_height']] = 0

        # Calculate instantaneous acceleration
        df['instant_acceleration'] = df['rate_of_change_height'] / df['time_diff']

        # Calculate direction in degrees
        df['direction_degrees'] = np.degrees(np.arctan2(df['rate_of_change_lat_mrcodegree'], df['rate_of_change_long_mrcodegree']))

        # Calculate rate of change of direction in degrees per second
        df['rate_of_change_direction_degrees'] = df['direction_degrees'].diff() / df['time_diff']

        # Handle cases where time difference is zero for direction
        df.loc[zero_time_diff_indices, 'rate_of_change_direction_degrees'] = 0

        # Save the processed DataFrame to a CSV file
        df.to_csv(output_file_path, index=False)

        print(f"Processed and saved file: {file_name}")
```

```python
import pandas as pd
import numpy as np

# Read the CSV file
file_path = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/updated_esp32_location_filtered_1.csv"
df = pd.read_csv(file_path)

# Convert 'Height' column from '90 m' format to numeric
df['Height'] = df['Height'].str.replace(' m', '').astype(float)

# Calculate latitude and longitude in radians
df['Latitude_radians'] = np.radians(df['Latitide'])
df['Longitude_radians'] = np.radians(df['Logitude'])

# Calculate time difference
df['time_diff'] = df['Fly Time (seconds)'].diff()

# Calculate rates using time difference
df['rate_of_change_lat_mrcodegree'] = (df['Latitude_radians'].diff() / df['time_diff']) * (10**6 * 180 / np.pi)
df['rate_of_change_long_mrcodegree'] = (df['Longitude_radians'].diff() / df['time_diff']) * (10**6 * 180 / np.pi)
df['rate_of_change_height'] = df['Height'].diff() / df['time_diff']

# Calculate instant acceleration
df['instant_acceleration'] = np.gradient(df['Speed Horizontal'], df['time_diff'])

# Make very small values zero
df.loc[df['instant_acceleration'].abs() < 1e-10, 'instant_acceleration'] = 0

# Calculate direction in degrees
df['direction_degrees'] = df['Direction']

# Calculate rate of change of direction in degrees
df['rate_of_change_direction_degrees'] = np.gradient(df['direction_degrees']) / df['time_diff']

# Rename columns
df.rename(columns={'Fly Time (seconds)': 'OSD.flyTime [s]',
                   'Latitide': 'OSD.latitude',
                   'Logitude': 'OSD.longitude',
                   'Height': 'OSD.height [m]',
                   'Speed Horizontal': 'OSD.hSpeed [m/s]',
                   'Latitude_radians': 'latitude_radians',
                   'Longitude_radians': 'longitude_radians',
                   'rate_of_change_lat_mrcodegree': 'rate_of_change_lat_mrcodegree',
                   'rate_of_change_long_mrcodegree': 'rate_of_change_long_mrcodegree',
                   'rate_of_change_height': 'rate_of_change_height',
                   'time_diff': 'time_diff',
                   'instant_acceleration': 'instant_acceleration',
                   'Direction': 'direction_degrees',
                   'rate_of_change_direction_degrees': 'rate_of_change_direction_degrees'}, inplace=True)

# Print the updated DataFrame
print(df)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVR
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error

# Read the CSV file
df = pd.read_csv("/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/updated_dataset.csv")



# Split data into features and target variable
X = df.drop(columns=['nature'])  # Assuming 'target' is the name of the target variable
y = df['nature']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Initialize and train models
linear_reg = LinearRegression()
linear_reg.fit(X_train_imputed, y_train)

random_forest = RandomForestRegressor()
random_forest.fit(X_train_imputed, y_train)

decision_tree = DecisionTreeRegressor()
decision_tree.fit(X_train_imputed, y_train)

gradient_boosting = GradientBoostingRegressor()
gradient_boosting.fit(X_train_imputed, y_train)

naive_bayes = GaussianNB()
naive_bayes.fit(X_train_imputed, y_train)

# svm_linear = SVR(kernel='linear')
# svm_linear.fit(X_train_imputed, y_train)

svm_rbf = SVR(kernel='rbf')
svm_rbf.fit(X_train_imputed, y_train)
```

```python
from sklearn.metrics import accuracy_score

# Initialize models
models = {
    'Linear Regression': linear_reg,
    'Random Forest': random_forest,
    'Decision Tree': decision_tree,
    'Gradient Boosting': gradient_boosting,
    'Naive Bayes': naive_bayes,
    'SVM RBF Kernel': svm_rbf
}

# Convert nature to binary labels if necessary
y_test_binary = (y_test > 0.5).astype(int)

# Calculate accuracy percentages
for name, model in models.items():
    y_pred = model.predict(X_test_imputed)
    y_pred_binary = (y_pred > 0.5).astype(int)
    accuracy = accuracy_score(y_test_binary, y_pred_binary)
    print(f"{name}: Accuracy = {accuracy * 100:.2f}%")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error

# Read the CSV file
df = pd.read_csv("/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/updated_dataset.csv")

# Split data into features and target variable
X = df.drop(columns=['nature'])
y = df['nature']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Initialize models
models = {
    'Linear Regression': (LinearRegression(), {'fit_intercept': [True, False]}),
    'Decision Tree': (DecisionTreeRegressor(), {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4]}),
    'Gradient Boosting': (GradientBoostingRegressor(), {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 10], 'learning_rate': [0.01, 0.1, 0.5]}),
    'SVR RBF Kernel': (SVR(kernel='rbf'), {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 'scale', 'auto']}),
}

# Perform hyperparameter tuning
results = {}
for name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, cv=3, n_jobs=-1, verbose=2)
    grid_search.fit(X_train_imputed, y_train)
    best_params = grid_search.best_params_
    print(f"Best Hyperparameters for {name}: {best_params}")

    # Evaluate the best model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test_imputed)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error for {name}: {mse}")

    results[name] = {'Best Hyperparameters': best_params, 'Mean Squared Error': mse}

# Print results
for name, metrics in results.items():
    print(f"{name}:\n{metrics}\n")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error

# Read the CSV file
df = pd.read_csv("/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/updated_dataset.csv")

# Split data into features and target variable
X = df.drop(columns=['nature'])
y = df['nature']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Handle missing values
imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

# Initialize models
models = {
    'Linear Regression': (LinearRegression(), {'fit_intercept': [True, False]}),
    'Decision Tree': (DecisionTreeRegressor(), {'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], 'min_samples_leaf': [1, 2, 4]}),
    'Gradient Boosting': (GradientBoostingRegressor(), {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 10], 'learning_rate': [0.01, 0.1, 0.5]}),
    'SVR RBF Kernel': (SVR(kernel='rbf'), {'C': [0.1, 1, 10], 'gamma': [0.1, 1, 'scale', 'auto']}),
}

# Perform hyperparameter tuning
results = {}
for name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, cv=3, n_jobs=-1, verbose=2)
    grid_search.fit(X_train_imputed, y_train)
    best_params = grid_search.best_params_
    print(f"Best Hyperparameters for {name}: {best_params}")

    # Evaluate the best model
    best_model = grid_search.best_estimator_
    y_pred = best_model.predict(X_test_imputed)
    mse = mean_squared_error(y_test, y_pred)
    print(f"Mean Squared Error for {name}: {mse}")

    results[name] = {'Best Hyperparameters': best_params, 'Mean Squared Error': mse}

# Print results
for name, metrics in results.items():
    print(f"{name}:\n{metrics}\n")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, LayerNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import Mean
import tensorflow as tf

# Load the CSV file into a DataFrame
file_path = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/combined_records_1_without_lat_long_sequential.csv"
df = pd.read_csv(file_path)

# Convert 'OSD.flyTime [s]' column to a numpy array
data = df['OSD.flyTime [s]'].values

# Normalize the data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data.reshape(-1, 1))

# Create sequences for input data
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 10  # Define the length of sequences
X, y = create_sequences(data_normalized, seq_length)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Build the transformer model
def transformer_model(seq_length, d_model=128, num_heads=2, dense_units=128, dropout_rate=0.2):
    inputs = Input(shape=(seq_length, 1))
    x = inputs

    # Transformer Encoder
    for _ in range(2):
        x = tf.keras.layers.MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(x, x)
        x = LayerNormalization(epsilon=1e-6)(x)
        x = Dropout(dropout_rate)(x)
        y = tf.keras.layers.Dense(dense_units, activation='relu')(x)
        y = tf.keras.layers.Dense(d_model)(y)
        x = tf.keras.layers.Add()([x, y])
        x = LayerNormalization(epsilon=1e-6)(x)
        x = Dropout(dropout_rate)(x)

    # Output layer
    x = tf.keras.layers.Flatten()(x)
    outputs = Dense(1)(x)

    model = Model(inputs=inputs, outputs=outputs)
    return model

model = transformer_model(seq_length)
model.compile(optimizer=Adam(), loss=MeanSquaredError(), metrics=[MeanSquaredError()])

# Train the model with data augmentation
for _ in range(10):
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1, validation_data=(X_test, y_test))
    # Add noise to the training data
    X_train_augmented = X_train + np.random.normal(0, 0.05, size=X_train.shape)
    model.fit(X_train_augmented, y_train, epochs=1, batch_size=32, verbose=1, validation_data=(X_test, y_test))

# Evaluate the model
mse = model.evaluate(X_test, y_test)
print(f"Mean Squared Error on Test Set: {mse}")
```

```python
# Calculate Accuracy
threshold = 0.3# Define a threshold for accuracy calculation
predictions = model.predict(X_test)
correct_predictions = np.mean(np.abs(predictions.flatten() - y_test.flatten()) <= threshold)
accuracy = correct_predictions * 100

print(f"Accuracy: {accuracy}%")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import Mean

# Load the CSV file into a DataFrame
file_path = "/content/drive/MyDrive/Saduvu biddalu/semesters/sem 8/major project 1/combined_records_1_without_lat_long_sequential.csv"
df = pd.read_csv(file_path)

# Convert 'OSD.flyTime [s]' column to a numpy array
data = df['OSD.flyTime [s]'].values

# Normalize the data
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data.reshape(-1, 1))

# Create sequences for input data
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 10  # Define the length of sequences
X, y = create_sequences(data_normalized, seq_length)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Build the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(seq_length, 1)))
model.add(Dense(1))

model.compile(optimizer=Adam(), loss=MeanSquaredError(), metrics=[MeanSquaredError()])

# Train the model with data augmentation
for _ in range(10):
    model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1, validation_data=(X_test, y_test))
    # Add noise to the training data
    X_train_augmented = X_train + np.random.normal(0, 0.05, size=X_train.shape)
    model.fit(X_train_augmented, y_train, epochs=1, batch_size=32, verbose=1, validation_data=(X_test, y_test))


# Evaluate the model
mse = model.evaluate(X_test, y_test)
print(f"Mean Squared Error on Test Set: {mse}")
```

```python
# Predict on test data
y_pred = model.predict(X_test)

# Inverse transform the normalized predictions and ground truth values
y_pred_inv = scaler.inverse_transform(y_pred)
y_test_inv = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate accuracy based on percentage of predictions within a certain range of actual values
range_threshold = 1  # Define the range threshold as a percentage of the actual value
within_range = np.abs(y_pred_inv - y_test_inv.squeeze()) / y_test_inv.squeeze() <= range_threshold
accuracy = np.mean(within_range) * 100
print(f"Accuracy: {accuracy:.2f}%")
```
